```python
# 把多个月份的BTC合约拼接到一起
def resample_data(file_path):
    z = pd.read_csv(file_path)  # 读取数据
    z['open_time'] = pd.to_datetime(z['open_time'], unit='ms').dt.strftime('%Y-%m-%d %H:%M:%S')  # 改变时间戳的格式，变成datetime格式
    z['close_time'] = pd.to_datetime(z['close_time'], unit='ms').dt.strftime('%Y-%m-%d %H:%M:%S')
    # 重置index，变成时间戳
    z.index = pd.to_datetime(z.open_time)
    # 降频处理
    z_rspled = z.resample('5min').agg({'open_time':'first', 
                              'open':'first', 
                              'high':'max', 
                              'low':'min', 
                              'close':'last', 
                              'volume':'sum', 
                              'close_time':'last',
                              'quote_volume':'sum', 
                              'count':'sum', 
                              'taker_buy_volume':'sum', 
                              'taker_buy_quote_volume':'sum',
                              'ignore':'last'})
    return z_rspled
    
```

```python
file_path_list = ["D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-01.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-02.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-03.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-04.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-05.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-06.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-07.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-08.zip",
                  "D:/workspace/data/crypto/1min/BTCUSDT/BTCUSDT-1m-2024-09.zip",]
```

```python
df_list = []
for i in file_path_list:
    z = resample_data(i)
    df_list.append(z)
```

```python
# 把9个dataframe按时间顺序拼接到一起
df = pd.concat(df_list, axis=0)
```
